###### 9.Stwórz przy pomocy matplotlib wykres rozrzutu (scatterplot) zgodnie z wytycznymi:
* oś x: indeksy
* wartości osi x: indeksy
* opis osi x: 'Major'
* oś y: mediana zarobków na początku kariery
* wartości osi y: jw
* opis osi y: 'US Dollars'
* tytuł: 'Starting Median Salary by Major'
* kolor obserwacji: zielony
* legenda: 'Starting Median Salary'

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df = dfs[0]

plt.scatter(df.loc[:,0:1], df['Starting Median Salary'], color = 'green')
plt.xscale('log')
xlab = 'Major'
ylab = 'US Dollars'
title = 'Starting Median Salary by Major'
x_skala_wartosci = [100, 1000, 10000, 100000]
x_skala_opis = ['Sto', '1 tysiąc', '10 tysięcy', '100 tysięcy']
plt.xticks(x_skala_wartosci, x_skala_opis)
y_skala_wartosci = [100, 1000, 10000, 100000]
y_skala_opis = ['Sto', '1 tysiąc', '10 tysięcy', '100 tysięcy']
plt.yticks(y_skala_wartosci, y_skala_opis)
plt.xlabel(xlab)
plt.ylabel(ylab)
plt.title(title)
plt.legend()
plt.show()
```
